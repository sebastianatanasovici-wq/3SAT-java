```python
import itertools
import time

def evaluar_clausula(clausula, asignacion):
    for lit in clausula:
        var = abs(lit)
        if (lit > 0 and asignacion[var] == 1) or (lit < 0 and asignacion[var] == 0):
            return True
    return False

def sat_bruteforce(formula, n):
    for valores in itertools.product([0, 1], repeat=n):
        asignacion = {i+1: valores[i] for i in range(n)}

        if all(evaluar_clausula(c, asignacion) for c in formula):
            print("Solución encontrada:")
            for i in range(1, n+1):
                print(f"x{i} = {asignacion[i]}")
            return True

    return False


# MISMA fórmula en todos
formula = [
    [1, -2, 3],
    [-1, 2, 4]
]
n = 4

inicio = time.time()
res = sat_bruteforce(formula, n)
fin = time.time()

print("La fórmula es SATISFACIBLE" if res else "La fórmula NO es satisfacible")
print("Tiempo de ejecución:", fin - inicio, "segundos")
```
